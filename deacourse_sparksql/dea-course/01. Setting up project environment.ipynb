```sql

-- Create external location on gizmobox
CREATE EXTERNAL LOCATION IF NOT EXISTS dea_course_ext_adls_gizmobox
  URL 'abfss://gizmobox@deacourseadlsa.dfs.core.windows.net/'
  WITH (STORAGE CREDENTIAL `dea-course-ext-sc`)
  COMMENT "gizmobox data for dea";
```

```sql
--CREATE CATALOGS for gizmobox 
CREATE CATALOG IF NOT EXISTS dea_course_catalog_gizmobox
COMMENT 'This is catalog for gizmobox datalake'
MANAGED LOCATION 'abfss://gizmobox@deacourseadlsa.dfs.core.windows.net/';
```

```sql
use catalog dea_course_catalog_gizmobox
```

```sql
--create schema for landing,bronze,gold,siver
CREATE SCHEMA IF NOT EXISTS dea_course_catalog_gizmobox.landing
COMMENT 'landing schema for raw files'
MANAGED LOCATION 'abfss://gizmobox@deacourseadlsa.dfs.core.windows.net/landing';

CREATE SCHEMA IF NOT EXISTS dea_course_catalog_gizmobox.bronze
COMMENT 'bronze schema for raw files'
MANAGED LOCATION 'abfss://gizmobox@deacourseadlsa.dfs.core.windows.net/bronze';

CREATE SCHEMA IF NOT EXISTS dea_course_catalog_gizmobox.silver
COMMENT 'silver schema for raw files'
MANAGED LOCATION 'abfss://gizmobox@deacourseadlsa.dfs.core.windows.net/silver';

CREATE SCHEMA IF NOT EXISTS dea_course_catalog_gizmobox.gold
COMMENT 'gold schema for raw files'
MANAGED LOCATION 'abfss://gizmobox@deacourseadlsa.dfs.core.windows.net/gold';
```

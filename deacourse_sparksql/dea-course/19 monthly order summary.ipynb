```sql
CREATE OR REPLACE TABLE dea_course_catalog_gizmobox.gold.order_summary_monthly AS
select
  customer_id,
  date_format(transaction_timestamp, 'yyyy-MM') as order_month,
  count(distinct order_id) as total_orders,
  sum(quantity) as total_items_bought,
  SUM(price * quantity) as total_amount
from
  dea_course_catalog_gizmobox.silver.orders
group by
  customer_id,
  order_month
```

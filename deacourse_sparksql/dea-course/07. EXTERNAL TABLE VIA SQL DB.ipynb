```sql
CREATE SCHEMA IF NOT EXISTS hive_metastore.bronze;
```

```sql
%python
dbutils.widgets.text("dbtable","")
dbutils.widgets.text("port","1433")
dbutils.widgets.text("schema","")
par_dbtable = dbutils.widgets.get("dbtable")
par_port = dbutils.widgets.get("port")
par_schema = dbutils.widgets.get("schema")
print(par_dbtable)
print(par_port)
print(par_schema)
```

used python cell to create variables to store secrets

Then used variables using spark.config.set to pass as sql parameters for connection string


```sql
%python

server = dbutils.secrets.get(scope="deacoursekeyvault-scope", key="deacoursesql-server")
db = dbutils.secrets.get(scope="deacoursekeyvault-scope", key="deacoursesql-database")
username = dbutils.secrets.get(scope="deacoursekeyvault-scope", key="deacoursesql-username")
password = dbutils.secrets.get(scope="deacoursekeyvault-scope", key="deacoursesql-pwd")
port = par_port
jdbc_url = f"jdbc:sqlserver://{server}:{port};database={db}"
spark.conf.set("jdbc.url", jdbc_url)
spark.conf.set("jdbc.username", username)
spark.conf.set("jdbc.password", password)
spark.conf.set("jdbc.dbtable", f"{par_schema}.{par_dbtable}")
```

```sql
DROP TABLE IF EXISTS hive_metastore.bronze.refunds ;
CREATE TABLE IF NOT EXISTS  hive_metastore.bronze.refunds
USING JDBC
OPTIONS (
  URL     '${jdbc.url}',
  dbtable '${jdbc.dbtable}',
  user    '${jdbc.username}',
  password '${jdbc.password}');
```

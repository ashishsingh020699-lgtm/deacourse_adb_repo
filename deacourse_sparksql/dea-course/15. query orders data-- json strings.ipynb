```sql
select * from dea_course_catalog_gizmobox.bronze.v_orders
```

To extract array elements: column_name:item_name


```sql
select 
value:order_id as order_id,
value:customer_id as customer_id
,value
from dea_course_catalog_gizmobox.bronze.v_orders v
```

extract array elements : column_name:item_array[index]


```sql
select 
value:order_id as order_id
,value:customer_id as customer_id
,value:items[0] as item_1
,value:items[1] as item_2
,value:items[2] as item_3
,value
from dea_course_catalog_gizmobox.bronze.v_orders v
```

Extract nested columns : column_name:item_array[index].leaf_node

```sql
select 
value:order_id as order_id
,value
,value:items[0].item_id as item0id
,value:items[0].category as item0category
,value:items[0].name as item0name
,value:items[0].price as item0price
,value:items[0].quantity as item0quantity
,value:items[0].quantity :: INT * value:items[0].price :: INT as item0total
,value:items[1].item_id as item1id
,value:items[1].category as item1category
,value:items[1].name as item1name
,value:items[1].price as item1price
,value:items[1].quantity as item1quantity
,value:items[1].quantity :: INT * value:items[1].price :: INT as item1total
,item1total + item0total as total
from dea_course_catalog_gizmobox.bronze.v_orders v
```

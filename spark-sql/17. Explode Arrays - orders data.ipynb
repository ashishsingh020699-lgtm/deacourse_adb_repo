access elements in json objects : column.object
<br> array_distinct : eliminates duplicate items in an array 


```sql
select  
    json_value.customer_id
    ,json_value.order_date
    ,json_value.order_id
    ,json_value.order_status
    ,json_value.payment_method
    ,json_value.total_amount    
    ,json_value.transaction_timestamp
    ,array_distinct(json_value.items) as distinct_items -- eliminates duplicate objects in array like order_id 29 , customer_id 5953
    ,json_value
 from dea_course_catalog_gizmobox.silver.orders
where json_value.order_id = 50 or json_value.order_id = 27
```

explode arrays using explode
<br>
each object in array will be represented by a single line

```sql
select  
    json_value.customer_id
    ,json_value.order_date
    ,json_value.order_id
    ,json_value.order_status
    ,json_value.payment_method
    ,json_value.total_amount    
    ,json_value.transaction_timestamp
    ,explode(array_distinct(json_value.items)) as item
 from dea_course_catalog_gizmobox.silver.orders
 where json_value.order_id = 50 or json_value.order_id = 27
```

```sql
CREATE OR REPLACE TEMPORARY VIEW tv_orders_exploded AS
select
  json_value.customer_id,
  json_value.order_date,
  json_value.order_id,
  json_value.order_status,
  json_value.payment_method,
  json_value.total_amount,
  json_value.transaction_timestamp,
  explode(array_distinct(json_value.items)) as item
from
  dea_course_catalog_gizmobox.silver.orders;
```

```sql
SHOW COLUMNS IN tv_orders_exploded
```

extracting nested objects in details from items

```sql
select
  customer_id,
  order_date,
  order_id,
  order_status,
  payment_method,
  total_amount,
  transaction_timestamp,
  item.item_id
  ,item.name
  ,item.price
  ,item.quantity
  ,item.category
  ,item.details.brand
  ,item.details.color
  ,item
from
  tv_orders_exploded
  where order_id = 50 or order_id = 27
```

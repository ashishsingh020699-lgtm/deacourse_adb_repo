
- we will use read_files function instead of just select so that we can overcome limitations such as 
- promoting headers 
- and mentioning delimiters


```sql
select * from csv.`/Volumes/dea_course_catalog_gizmobox/landing/operational_data/addresses/`
```

https://docs.databricks.com/aws/en/sql/language-manual/functions/read_files

```sql
-- USING read_files, we can provide additional parameters like format,header,delimiter
select
  *
from
  read_files(
    '/Volumes/dea_course_catalog_gizmobox/landing/operational_data/addresses/',
    format => 'csv',
    header => true,
    delimiter => '\t'
  )
```

```sql
CREATE OR REPLACE VIEW dea_course_catalog_gizmobox.bronze.v_addresses AS
Select
  *
from
  read_files(
    "/Volumes/dea_course_catalog_gizmobox/landing/operational_data/addresses/",
    header => true,
    delimiter => '\t',
    format => 'csv'
  )
```

```sql
select count(1) from dea_course_catalog_gizmobox.bronze.v_addresses 
```
